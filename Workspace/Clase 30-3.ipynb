```python
#
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
data = pd.read_excel("StockPrice.xlsx", sheet_name="StockPrice")
df = pd.DataFrame(data)

#SDT TEMP_AGUA Variables continuas 
sample = df.loc[:,["Google", "TOYOTA"]]
sample.head()

X = sample["Google"]
Y = sample["TOYOTA"]
##Metodos de numpy  omiten los NAN 

fig, ax = plt.subplots(figsize = (7,5))
ax.scatter(sample["Google"], sample["TOYOTA"]) 


#Regresion lineal 

covaXY = np.cov(X,Y)
b1 = covaXY[0][1]/np.var(X)
b1

b0 = np.mean(Y) - b1 *np.mean(X)
b0
#fn lineal

x = np.linspace(X.min(), X.max(), 100) # Intervalo 
y_est = b0 + b1 * x #<-- aprendizaje 

##Imprimir el aprendizaje
ax.plot(x,y_est, color="red")

#Etapa de consulta
y_pred = []
for i in range(len(X)):
    y_pred.append(b0 + b1 * X[i]) 
y_real = Y
#Metricas de evealuacion 
#Coef, ede determinacion es la correlacion al cuadrado
coefDet = pow(np.corrcoef(y_pred,y_real)[0][1],2)
coefDet
rmse = np.sqrt(sum(pow(y_pred-y_real,2))/len(Y))
rmse
```

```python
def linearRegresion(X,Y): 

    fig, ax = plt.subplots(figsize = (7,5))
    ax.scatter(sample["Google"], sample["TOYOTA"]) 


    #Regresion lineal 

    covaXY = np.cov(X,Y)
    b1 = covaXY[0][1]/np.var(X)
    b1

    b0 = np.mean(Y) - b1 *np.mean(X)
    b0
    #fn lineal

    x = np.linspace(X.min(), X.max(), 100) # Intervalo 
    y_est = b0 + b1 * x

    ##Imprimir el aprendizaje
    ax.plot(x,y_est, color="red")
    ax.set_title("Lienar regression statics")
    ax.set_xlabel("Empresa 1 ")
    ax.set_ylabel("Empresa 2 ")



X = sample["Google"]
Y = sample["TOYOTA"]

linearRegresion(X,Y)
```
